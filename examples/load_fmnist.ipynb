Example of Fashion-MNIST dataset loading 
---------------------------------------------------------------

```python
from ml_datasets.fmnist import FashionMNIST
from ml_datasets.utils import plot_images
```

The Fashion-MNIST dataset is downloaded from http:///
The first step is to initialize the FashionMNIST class. This class accepts several argumetns:
args: 
    x
    y
    z

```python
fmnist = FashionMNIST(verbose=False)
x_train, y_train, x_test, y_test = fmnist.load()

```

```python
print("x_train: {}, y_train: {}, x_test: {}, y_test: {}".format(x_train.shape, y_train.shape, x_test.shape, y_test.shape))
```

```python
labels = fmnist.meta
print(labels)
```

```python
num_sample_perclass = 10
title = "Fashion-MNIST Dataset ({} samples)".format(num_sample_perclass * len(labels))

plot_images(
    num_sample_perclass=num_sample_perclass,
    x=x_train,
    y=y_train,
    labels=labels,
    title=title,
    cmap="Greys_r",
    )
```
